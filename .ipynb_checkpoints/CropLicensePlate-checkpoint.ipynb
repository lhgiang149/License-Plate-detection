```python
import cv2
from matplotlib import pyplot as mlt
import numpy as np
from PIL import Image

def findLargestArea(longList):
    biggest_A = 0
    for contour in longList:
        x,y,w,h = cv2.boundingRect(contour)
        if x==0 and y==0:
            continue
        A = w*h
        if (A > biggest_A):
            license_plate = (x,y,w,h)
            biggest_A = A
    return license_plate
def findLicensePlate(img_path):
    '''
    Input:
    + img_path: path of image in your computer
    Output: an 100x100 pixels RGB picture of license plate in your picture
    '''
    image = cv2.imread(img_path)
    gray =  cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#     ret, img = cv2.threshold(gray,150,255,cv2.THRESH_BINARY_INV)
    ret,img = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
#     mlt.imshow(img,cmap = 'gray')
    kernel = np.ones((11,11),np.uint8)
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
#     mlt.imshow(closing, cmap ='gray')
    contours, hierarchy = cv2.findContours(closing,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    temp = findLargestArea(contours)
    license_loc = image[temp[1]:(temp[1]+temp[3]),temp[0]:(temp[0]+temp[2])]
    print(license_loc.shape)
    image = Image.fromarray(license_loc)
    liPlate = image.resize((100, 100), Image.BILINEAR) 
    return liPlate
license = findLicensePlate('M:\\Project2_data\\GreenParking\\0000_00532_b.jpg')
# image = Image.fromarray(license)
# im2 = image.resize((100, 100), Image.BILINEAR)
mlt.imshow(license)
# mlt.imshow(findLicensePlate('2.jpg'))
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

license_cascade = cv2.CascadeClassifier('vn2.xml')

img_path = 'M:\\Project2_data\\GreenParking\\'

count = '1'
count_str = 1

def findLargestArea(longList):
    biggest_A = 0
    for contour in longList:
        x,y,w,h = contour
        A = w*h
        if (A > biggest_A):
            license_plate = np.array([[x,y,w,h]])
            biggest_A = A
    return license_plate


for i in os.listdir(img_path):
    image = cv2.imread(img_path+i)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    license = license_cascade.detectMultiScale(gray, 1.3,5)
    if(license == ()):
        continue
    license = findLargestArea(license)
    img = image[license[0,1]:(license[0,1]+license[0,3]),license[0,0]:(license[0,0]+license[0,2])]
    cv2.imwrite(('M:\\Project2_data\\crop_by_cascade\\'+count+'.jpg'), img) 
    count_str += 1
    count = str(count_str)

    
```

```python
video_path = 'C:\\Users\\ADMINS\\Desktop\\parking.mp4'
license_cascade = cv2.CascadeClassifier('vn2.xml')

cap = cv2.VideoCapture(video_path)
video = []
count = 0 
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        continue
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    license = license_cascade.detectMultiScale(gray, 3,6)
    if len(license)!=0:
        img = frame[license[0,1]:(license[0,1]+license[0,3]),license[0,0]:(license[0,0]+license[0,2])]
        video.append(img)
        count +=1
    if count == 3:
        break

cap.release()
cv2.destroyAllWindows()
```
